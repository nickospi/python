```python
import requests
from bs4 import BeautifulSoup
import re
from urllib.request import urlopen
import pandas as pd
import numpy as np
import json
```

```python
raw_html = urlopen("http://floatingmedia.com/columbia/BBC.html").read()
soup_doc = BeautifulSoup(raw_html, "html.parser")
print(soup_doc.prettify())
```

```python
all_info = soup_doc.find(class_='body-content') 
all_p = all_info.findAll('p')[1:] 
```

```python
list_of_movies = []

for lines in all_p:
    if lines.strong is not None:
        critic_info = lines.find('strong').string
        regex_for_name = r"^\w*.\w*" 
        critic_name = re.findall(regex_for_name,critic_info)
        critic_name = ", ".join(critic_name)
        regex_for_org = r"(?:–)(\s\w*'?-?\.?\s?\w*’?\.?-?\s?\w*\s?\w*)"
        critic_org = re.findall(regex_for_org,critic_info)
        critic_org = ", ".join(critic_org)
        regex_for_cn = r"(?:\()(\w*\s?\w*?)(?:\))"
        critic_cn = re.findall(regex_for_cn,critic_info)
        critic_cn = ", ".join(critic_cn)
        movie_info = lines.findNextSibling('p')
        each_movie = movie_info.find_all(string=True)
        for movie in each_movie: 
            m_data = {}
            m_data ['critic_name'] = critic_name
            m_data ['critic_org'] = critic_org
            m_data ['critic_cn'] = critic_cn
            regex_for_rank = r"^\d*\b" 
            movie_rank = re.findall(regex_for_rank,movie)
            m_data['movie_rank'] = ", ".join(movie_rank)
            regex_for_movie_name = r"(?:\.\s)(?:\w*?:?,?!?'?’?\w?\d*?-?\w*?\s)(?:\d*?\w*?’?:?,?\w?\s)?(?:\w*?:?!?\d*?\s)?(?:\w*?\s?&?\s)?(?:\d*?\w*?\s)?(?:\w*,?\s)?(?:\w*\s)?(?:\w*\s)?(?:assassino gentile)?(?:\w*\s)?(?:\w*\s)?(?:\w*\??\s)?" 
            movie_name = re.findall(regex_for_movie_name,movie)
            m_data['movie_name'] = ", ".join(movie_name)
            regex_for_dir = r"(?:\()(?:\w*-?\s?\w*?\.?\s?)(?:and\s)?(?:\w*?\s?\.?-?,?\w*?\s?)?(?:and\s)?(?:\w*?-?\s?)?(?:and\s)?(?:\w*?-?\s?)(?:\w*?-?\s?)(?:,)"
            movie_dir = re.findall(regex_for_dir,movie)
            m_data['movie_dir'] = ", ".join(movie_dir)
            regex_for_yr = r"2\d{3}"
            movie_yr = re.findall(regex_for_yr,movie)
            m_data['movie_yr'] = ", ".join(movie_yr)
            list_of_movies.append(m_data)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
driver = webdriver.Chrome()
```

```python
#Scraping imdb.com for director's country/age
#Getting director's page url from search results url
base_url = 'https://www.imdb.com/find?ref_=nv_sr_fn&q='
for item in df_csv_movies.movie_dir:
    def get_dir_url(row):
        driver.get(base_url+row['movie_dir'])
        try: #no 'ands'         
            dir_page = driver.find_element_by_class_name('result_text').find_element_by_tag_name ('a').get_attribute('href')
        except:
            dir_page = 'no_page'


        return pd.Series({
            'dir_page': dir_page
        })
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```
